```r
timestosample <- 10000 #number of datasets to sample from the population

bimodal <- function(size){
  ifelse(sample(0:1,size,replace = TRUE),rnorm(size,0,1),rnorm(size,10,4))
}

samplesize <- 5 #size of EACH sample set
sampleaverages <- vector(mode="numeric", length=timestosample)
for (i in 1:timestosample){
    a <- bimodal(samplesize)
    sampleaverages[i] <- mean(a)
}

df <- data.frame(age=factor(rep("Sample average",each=timestosample)),sample_averages=c(sampleaverages))

library(ggplot2)
ggplot(df, aes(x=sample_averages)) +  geom_histogram(binwidth=.3,color="black", fill="white") + xlim(-8,24) + coord_fixed(ratio=.05)
```
