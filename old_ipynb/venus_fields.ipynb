### Part B: Do different regions of Venus exhibit different magnetic fields? Do a t-test first.

```r
means=c()
CIs=c()
for (i in 1:max(VEX_data$GROUP)){
    means <- c(means,mean(VEX_data[VEX_data[,1] == i,]$Br.nT.))
    CIs <- c(CIs,sd(VEX_data[VEX_data[,1] == i,]$Br.nT.)/(nrow(VEX_data[VEX_data[,1] == 1,]))^(1/2))
}
dfmeans <- data.frame(means=means,
                  CIs=CIs)
dfmeans
```
