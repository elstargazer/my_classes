```r
#needed to make plots
library(ggplot2)
```

```r
#read in the data files. 
GPS1 <- read.csv(file = '/Users/rogerfu/Dropbox/my_classes/EPS_102_2021/Rogers_lectures/Lecture_0/GPS1.csv', stringsAsFactors = FALSE)
GPS2 <- read.csv(file = '/Users/rogerfu/Dropbox/my_classes/EPS_102_2021/Rogers_lectures/Lecture_0/GPS2.csv', stringsAsFactors = FALSE)
GPS3 <- read.csv(file = '/Users/rogerfu/Dropbox/my_classes/EPS_102_2021/Rogers_lectures/Lecture_0/GPS3.csv', stringsAsFactors = FALSE)
GPS4 <- read.csv(file = '/Users/rogerfu/Dropbox/my_classes/EPS_102_2021/Rogers_lectures/Lecture_0/GPS4.csv', stringsAsFactors = FALSE)
GPS5 <- read.csv(file = '/Users/rogerfu/Dropbox/my_classes/EPS_102_2021/Rogers_lectures/Lecture_0/GPS5.csv', stringsAsFactors = FALSE)
```

```r
#combine groups of data into one dataframe
df <- data.frame(group=c(rep("group 1",nrow(GPS1)),rep("group 2",nrow(GPS2)),rep("group 3",nrow(GPS3)),rep("group 4",nrow(GPS4)),rep("group 5",nrow(GPS5))),
                  latitude=Map(c, GPS1[2], GPS2[2],GPS3[2],GPS4[2],GPS5[2])   )
#total number of data points
nrow(df)
ggplot(df, aes(x=latitude,fill=group)) + geom_histogram( color="#e9ecef", alpha=0.6, position = 'identity') +
    scale_fill_manual(values=c("#69b3a2", "#404080", "#00CCCC", "#FF8000", "#CC0000")) +
    coord_fixed(ratio=.00003)
# RGB colors: https://www.rapidtables.com/web/color/RGB_Color.html
```
