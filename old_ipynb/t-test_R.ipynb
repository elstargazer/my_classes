## Latitude data for the Pacific Plate

```r
lats160 <- c(3.2, 7.8, 11.2,  -3.5, 5.3, 3.1, -2.7, 3.5, 7.2, 10.9, 9.2, 6.6, -2.1,
5.6, 3.8, -2.9, -0.6, -0.2, 6.9, 0.8)
lats145 <- c(31.9, 0.4, 0.7, 1.8, 2.5, 17.7, 31., 14., 5.4, 16.4, 10.3, 11.5)
```

```r
t.test(lats160,y=lats145)
```

```r
PacificPlateLatitudes <- data.frame(
    age = c(rep("160 Ma",length(lats160)),rep("145 Ma",length(lats145))),
    latitude = c(lats160,lats145)
)

PacificPlateLatitudes

library(ggplot2)
ggplot(PacificPlateLatitudes, aes(x=latitude, color=age)) +  geom_histogram(binwidth=5, position="identity",fill="white", alpha=0.3,linetype="solid")+
  theme(legend.position="top") + coord_fixed(ratio=1)
```

```r
#set.seed(1234)

df <- data.frame(
  age=factor(c(rep("160 Ma",each=20), rep("145 Ma",each=12))),
  latitude=round(c(rnorm(20, mean=3.6, sd=4), rnorm(12, mean=12, sd=8)),digits=1)
  )
ggplot(df, aes(x=latitude, color=age)) +  geom_histogram(binwidth=4, position="identity",fill="white", alpha=0.5)+
  theme(legend.position="top")
    }
```

```r
#set.seed(1234)
df <- data.frame(
  age=factor(c(rep("160 Ma",each=20), rep("145 Ma",each=12))),
  latitude=round(c(rnorm(20, mean=3.6, sd=3), c(rnorm(12/2, mean=6, sd=5),rnorm(12/2, mean=18, sd=5))),digits=1)
  )
head(df)
ggplot(df, aes(x=latitude, color=age)) +  geom_histogram(binwidth=2, position="identity",fill="white", alpha=0.5)+
  theme(legend.position="top")
```

```r
timestosample <- 1000 #number of datasets to sample from the population

samplesize160 <- .5*20 #size of EACH sample for the 160 Ma dataset
sampleaverages160 <- vector(mode="numeric", length=timestosample)
for (i in 1:timestosample){
    a <- rnorm(samplesize160, mean=3.6, sd=3)
    sampleaverages160[i] <- mean(a)
}

samplesize145 <- .5*12
sampleaverages145 <- vector(mode="numeric", length=timestosample)
for (i in 1:timestosample){
    a <- rnorm(samplesize145, mean=12, sd=5)
    sampleaverages145[i] <- mean(a)
}

df <- data.frame(age=factor(c(rep("160 Ma",each=timestosample), rep("145 Ma",each=timestosample))),sample_average=c(sampleaverages160,sampleaverages145))

ggplot(df, aes(x=sample_average)) +  geom_histogram(binwidth=.3,color="black", fill="white") + xlim(-5,25) + coord_fixed(ratio=.04)
```
