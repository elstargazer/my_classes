## Paleolatitude data from Pacific Plate

```r
lats160 <- c(3.2, 7.8, 11.2,  -3.5, 5.3, 3.1, -2.7, 3.5, 7.2, 10.9, 9.2, 6.6, -2.1,
5.6, 3.8, -2.9, -0.6, -0.2, 6.9, 0.8)
lats145 <- c(31.9, 0.4, 0.7, 1.8, 2.5, 17.7, 31., 14., 5.4, 16.4, 10.3, 11.5)
```

```r
PacificPlateLatitudes <- data.frame(
    age = c(rep("160 Ma",length(lats160)),rep("145 Ma",length(lats145))),
    latitude = c(lats160,lats145)
)

library(ggplot2)
ggplot(PacificPlateLatitudes, aes(x=latitude, color=age)) +  geom_histogram(binwidth=5, position="identity",fill="white", alpha=0.3,linetype="solid")+
  theme(legend.position="top") #+ coord_fixed(ratio=1)
```

```r
timestosample <- 500 #number of datasets to sample from the population

bimodal <- function(size){
  ifelse(sample(0:1,size,replace = TRUE),rnorm(size,0,1),rnorm(size,10,4))
}

samplesize <- 10 #size of EACH sample for the 160 Ma dataset
sampleaverages <- vector(mode="numeric", length=timestosample)
for (i in 1:timestosample){
    a <- bimodal(samplesize)
    sampleaverages[i] <- mean(a)
}

df <- data.frame(age=factor(rep("Sample average",each=timestosample)),single_measurements=c(sampleaverages))

ggplot(df, aes(x=single_measurements)) +  geom_histogram(binwidth=.3,color="black", fill="white") + xlim(-8,24) + coord_fixed(ratio=.02)
```
